```python
import requests
import pandas as pd
import time
import json
import os
from requests.adapters import HTTPAdapter, Retry
```

```python
#Leitura do json com todas as marcas
df_marcas = pd.read_json("marcas/marcas.json")
```

```python
#Cria diretório caso não exista
path_modelos = "modelos"

if not os.path.isdir(path_modelos):
    os.makedirs(path_modelos)
```

```python
for linha in df_marcas.index:

    #Obtém código da marca
    codigo_marca = df_marcas.loc[linha, "codigo"]

    #Cria url dinâmica
    url_modelo = f"https://parallelum.com.br/fipe/api/v1/carros/marcas/{codigo_marca}/modelos"

    #Cria dicionário com código da marca
    codigo_marca_dict = {"codigo_marca": codigo_marca}

    #Realiza requisição e converte para json
    modelos = requests.get(url_modelo, headers={"User-Agent": "XY"})
    modelos = modelos.json()

    #Atualiza o json com o código da marca
    for modelo in modelos["modelos"]:
        modelo.update(codigo_marca_dict)

    #Converte o dicionário para json
    modelos_json = json.dumps(modelos["modelos"], default=str)

    #Cria o nome do arquivo
    nome_arquivo = f"modelos_marca_{codigo_marca}.json"

    #Salva o arquivo json e fecha a instância aberta
    json_file = open(f"{path_modelos}/{nome_arquivo}", "w")
    json_file.write(modelos_json)
    json_file.close()

    #print("Arquivo {} salvo com sucesso.".format(nome_arquivo))

    time.sleep(1)
```
