```python
import requests
import pandas as pd
import time
import json
import os
from glob import glob
from requests.adapters import HTTPAdapter, Retry
```

```python
path_modelos = "modelos/"
```

```python
#Leitura do json com todos os modelos
arquivos = glob(os.path.join(path_modelos, "*.json"))
ind_df = (pd.read_json(f) for f in arquivos)
df_modelos = pd.concat(ind_df, ignore_index=True)
```

```python
#Cria diretório caso não exista
path_anos = "anos"

if not os.path.isdir(path_anos):
    os.makedirs(path_anos)
```

```python
#Listando os anos já coletados
lista_anos = os.listdir('anos')
```

```python
for linha in df_modelos.index:
    
    #Obtém os códigos necessários
    codigo_marca = df_modelos.loc[linha, "codigo_marca"]
    codigo_modelo = df_modelos.loc[linha, "codigo"]

    #Define a url e nome do arquivo
    url_anos = f"https://parallelum.com.br/fipe/api/v1/carros/marcas/{codigo_marca}/modelos/{codigo_modelo}/anos"
    nome_arquivo = f"anos_marca_{codigo_marca}_modelo_{codigo_modelo}.json"

    if nome_arquivo not in lista_anos:

        #Realiza o request na url dos anos
        anos = requests.get(url_anos, headers={"User-Agent": "XY"})
        anos = anos.json()

        #Cria o dicionário com o código da marca e código do modelo
        dict_codigo_marca_modelo = {"codigo_marca": codigo_marca, "codigo_modelo": codigo_modelo}

        #Atualiza o json com o código da marca
        for ano in anos:
            ano.update(dict_codigo_marca_modelo)
    
        #Converte o dicionário para json
        anos_json = json.dumps(anos, default=str)

        #Salva o arquivo json e fecha a instância aberta
        json_file = open(f"{path_anos}/{nome_arquivo}", "w")
        json_file.write(anos_json)
        json_file.close()

        print("Arquivo {} salvo com sucesso.".format(nome_arquivo))

        time.sleep(1)
```
